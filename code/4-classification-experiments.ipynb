```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import classification_report
import numpy as np
from scipy.special import expit
```

# Logistic regression
classification model activated by a exp equation

```python
_data = make_blobs(1000, 2, centers=2, random_state=2610, cluster_std=2)
```

```python
plt.scatter(_data[0][:, 0], _data[0][:, 1], c=_data[1])
```

```python
df = pd.DataFrame(_data[0], columns=['x1', 'x2'])
df['y'] = _data[1]
df.head()
```

```python
log_reg_model = LogisticRegression()
log_reg_model.fit(df[['x1', 'x2']], df['y'])
```

```python
y_pred = log_reg_model.predict(df[['x1', 'x2']])
```

```python
report = classification_report(df['y'], y_pred)
print(report)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-15,5,1000)
y=(0.9617/1.52122)*x-0.31953269

todos_lo_puntos=[]

for x_pun in np.linspace(-15,5,100):
    for y_pun in np.linspace(-15,5,100):
        todos_lo_puntos.append([x_pun,y_pun])
todos_lo_puntos=np.array(todos_lo_puntos)
y_pred_tod= log_reg_model.predict(todos_lo_puntos)

plt.scatter(_data[0][:,0],_data[0][:,1],c=_data[1])
plt.scatter(todos_lo_puntos[:,0],todos_lo_puntos[:,1],c=y_pred_tod,alpha=0.1)
plt.plot(x,y)
plt.show()
```
